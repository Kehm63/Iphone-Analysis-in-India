# Iphone Sales Analysis

```python
# import libraries
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
# reading the needed csv file with pandas and checking the number of null values
data=pd.read_csv('apple_products.csv')
data.isnull().sum()
```

```python
data.describe()
```

## Data sorting

All the data about the top 10 highest rated Iphones will be stored in another dataframe in order to provide more clarity on the most appreciated Iphones in India.

```python
top_rated= data.sort_values(by=['Star Rating'],ascending=False)
top_rated=top_rated.head(10)
```

```python
print(top_rated['Product Name'])
```

 According to 'Star rating', the above are the most appreciated Iphones in India
 

## Visualizations

```python
figure= px.bar(top_rated,y=top_rated['Number Of Ratings'],x=top_rated['Product Name'],title='number of ratings')
figure.show(renderer='png') 
```

The above shows the top rated phones alongside their number of ratings

### Relationship between Sale Price and Number of Ratings

```python
figure1=px.scatter(data_frame = data,trendline='ols', x="Number Of Ratings", y="Sale Price",title="Relationship between Sale Price and Number of Ratings of iPhones")
figure1.show(renderer='png')
```

The relationship between sale price and number of ratings is a negative linear one. This means Iphones with lower prices tend to sell more in India and vice versa

### Relationship between Discount Percentage and Number of Ratings 

```python
figure2=px.scatter(data_frame = data, x="Number Of Ratings",trendline='ols', y="Discount Percentage", size="Sale Price",title="Relationship between Discount Percentage and Number of Ratings of iPhones")
figure2.show(renderer='png')
```

There's a positive linear relationship between discount percentage and number of ratings. According to the scatterplot above, phones with a higher discount percentage sold more.
